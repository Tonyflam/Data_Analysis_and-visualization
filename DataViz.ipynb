```python
covid.columns = covid.columns.str.replace('/', '')
```

```python
covid.columns = covid.columns.str.lower()
```

```python
pop_per_cont= covid.groupby(['continent'])['population'].agg(['sum', 'mean'])
```

```python
pop_per_cont.reset_index(inplace=True)
pop_per_cont.plot(x='continent', y = ['sum', 'mean'], kind='bar')
```

```python
pop_per_cont.plot(x='continent', y ='mean', kind='barh')
```

```python
pop_per_cont.sort_values(by='mean', ascending=False, inplace = True)
pop_per_cont.plot(x='continent', y ='mean', kind='barh')
```
